```python
import xarray as xr
import geopandas as gpd
import pandas as pd
```

```python
# load the station data and the weather data
station_gpd = gpd.read_file('MonitorStation\matching_stations.shp')
weather_ds = xr.open_mfdataset('downloaded_files/GEOSFP.*.nc', combine='by_coords', chunks= None)
```

```python
# load the traffic data
traffic_jan = pd.read_csv('jan_allstates.csv')
traffic_feb = pd.read_csv('feb_allstates.csv')
traffic_mar = pd.read_csv('mar_allstates.csv')
```

```python
# combine the traffic data and filter the data for Utah
traffic_data = pd.concat([traffic_jan, traffic_feb, traffic_mar], axis=0)
traffic_data = traffic_data[traffic_data['State_Code'] == 49]
traffic_data
```

```python
print(station_gpd)
```

```python
## Extract the last 3 digits of the station id since the station id in the traffic data is different
station_gpd['Station_Id'] = station_gpd['Station_Id'].astype(str).str[-3:]
print(station_gpd.head())
```

```python
## Merge the traffic data with the station data based on the station id
traffic_data['Station_Id'] = traffic_data['Station_Id'].astype(str)
station_gpd['Station_Id'] = station_gpd['Station_Id'].astype(str)
merged_data = pd.merge(traffic_data, station_gpd, on='Station_Id', how='left')
print(merged_data.isnull().sum())
```

```python
## Save the merged data to a csv file
merged_data.to_csv('merged_data.csv', index=False)
```
